```python
import pymongo
from pymongo import MongoClient

#1.链接本地数据库服务
name = MongoClient('localhost')
#2.链接本地数据库
db = name.python_test   
#3.创建，连接集合
rec = db.record  
```

```python
import pandas as pd 
data = pd.DataFrame(list(rec.find()))
# data
```

```python
score = data[["total"]]
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450,500,550,600,650,700]
score2 = pd.cut(score.values.flatten(), bins=bins)
score2
```

```python
score2 = pd.DataFrame(score2, columns=["status"]) 
score2
```

```python
data["status"] = score2
data
```

```python
# # 每个区间学生的总人数
data_status_count = data.pivot_table(
    index="status",
    values="total",
    aggfunc={"total":"count"}
)
data_status_count.rename(columns={"total": "总人数"},inplace=True)
data_status_count
```

```python
# # 每个区间学生总分的均值
data_status_mean = data.pivot_table(
    index="status",
    values="total",
    aggfunc={"total":"std"}
)
data_status_mean.rename(columns={"total": "均值"},inplace=True)
data_status_mean
```

```python
# # 每个区间学生总分的标准差
data_status_std = data.pivot_table(
    index="status",
    values="total",
    aggfunc={"total":"std"}
)
data_status_std.rename(columns={"total": "标准差"},inplace=True)
data_status_std
```

```python
# # 每个区间学生总分的方差
data_status_var = data.pivot_table(
    index="status",
    values="total",
    aggfunc={"total":"var"}
)
data_status_var.rename(columns={"total": "方差"},inplace=True)
data_status_var
```

```python
index_status = pd.DataFrame(['(0, 50]', '(50, 100]',"(100, 150]","(150, 200]","(200, 250]","(250, 300]","(300, 350]","(350, 400]","(400, 450]","(450, 500]","(500, 550]","(550, 600]","(600, 650]","(650, 700]"])
index_status
```

```python
output = pd.concat([data_status_count,data_status_mean,data_status_var],axis=1)
output
```

```python
# output.to_csv("人数-均值-方差.csv")
```

```python
output_df = output.to_dict("records")
output_df
```

```python
data_rank = data_56_cn.sort_values("math",ascending=False)
data_rank.drop_duplicates()
```
